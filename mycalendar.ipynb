```python
def day_in_month(month,year):
    oddmonths=[1,3,5,7,8,10,12]
    evenmonths=[2,4,6,9,11]
    if month in oddmonths:
        return 31
    else:
        if month==2:
            if (year%4==0 or year%100==0) and year%400!=0:
                return 29
            else:
                return 28
        else:
            return 30

            
import datetime
import turtle
t=turtle.Turtle()
t.pensize(2)
t.speed(7)

dt = datetime.datetime.today()
year=dt.year
month=dt.month
day=dt.day

# Find first weekday of a month
weekday = datetime.datetime(year, month, 1).weekday()
if weekday==0:
    startday="Mon"
elif weekday==1:
    startday="Tue"
elif weekday==2:
    startday="Wed"
elif weekday==3:
    startday="Thurs"
elif weekday==4:
    startday="Fri"
elif weekday==5:
    startday="Sat"
else:
    startday="Sun"

# Draw rectangle
t.penup()
t.goto(-175,200)
t.pendown()
t.forward(175)
t.right(90)
t.forward(200)
t.right(90)
t.forward(175)
t.right(90)
t.forward(200)
t.right(90)
t.goto(-175,175)

# Draw grids
# 1.Vertical lines
for i in range(3):
    t.right(90)
    t.forward(175)
    t.left(90)
    t.forward(25)
    t.left(90)
    t.forward(175)
    t.right(90)
    t.forward(25)

# 2.Title section
t.right(90)
t.forward(175)
t.goto(-175,0)
t.left(90)

# 3.Horizontal lines
for i in range(4):
    t.forward(175)
    t.left(90)
    t.forward(25)
    t.left(90)
    t.forward(175)
    t.right(90)
    t.forward(25)
    t.right(90)

# Write title
t.penup()
t.goto(-120,180)
t.pendown()
t.write("{0} {1}".format(datetime.datetime(year, month, day).strftime("%B"),year))
t.penup()

# Write weekdays
weekday_list = ["Su","Mo","Tu","We","Th","Fr","Sa"]

for i in range(len(weekday_list)):
    t.goto(-170+25*i,160)
    t.pendown()
    t.write(weekday_list[i])
    t.penup()
    t.goto(-170+25*i,160)

# Write dates for the first week
a=0
if startday=="Mon":
    for i in range(1,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
elif startday=="Tue":
    for i in range(2,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
elif startday=="Wed":
    for i in range(3,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
elif startday=="Thurs":
    for i in range(4,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
elif startday=="Fri":
    for i in range(5,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
elif startday=="Sat":
    for i in range(6,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
else:
    for i in range(0,7):
        a+=1
        t.penup()
        t.goto(-170+25*i,135)
        t.pendown()
        t.write(a)
        
# Write dates for the rest of the month
for i in range(1,6):
    t.penup()
    t.goto(-170,135-25*i)
    for j in range(7):
        a+=1
        t.penup()
        t.goto(-170+25*j,135-25*i)
        t.pendown()
        if a>day_in_month(month,year):
            a=1
            t.write(a)            
        else:
            t.write(a)     
        
turtle.done()
```
