```python
from tkinter import *
import tkinter as tk
import math

text_display=""
first_number=0
second_number=0
def number_pressed(event):
    global text_display
    pressed=event.widget["text"]
    if pressed!="0" or (pressed=="0" and text_display!=""):
        text_display+=pressed
    else:
        text_display=""
    display()
    
def display():
    screen.delete(0,"end")
    if text_display=="":
        screen.insert(0,"0")
    else:
        screen.insert(0,text_display)
        
def action_pressed(event):
    global first_number
    first_number=float(screen.get())
    global text_display
    text_display=""
    global action
    action=event.widget["text"]
    
def finish_pressed(event):
    global second_number
    global first_number
    global action
    second_number=float(screen.get())
    if action=="+":
        result=first_number+second_number
    elif action=="*":
        result=first_number*second_number
    elif action=="-":
        result=first_number-second_number
    elif action=="/":
        if second_number!=0:
            result=first_number/second_number
        else:
            result="ERROR"
    else:
        result="ERROR"
    screen.delete(0,"end")
    screen.insert(0,result)
    global dec
    dec=False
    global text_display
    text_display=""
    action=""
    
def clear(event):
    global text_display
    text_display=""
    global first_number
    first_number=0
    global second_number
    second_number=0
    global action
    action=""
    display()
    
def clear_e(event):
    global text_display
    action=event.widget["text"]
    if action=="":
        clear(event)
    else:
        text_display=""
        display()
        
def backspace(event):
    global text_display
    if text_display=="":
        return 
    text_display=screen.get()
    text_display=text_display[:-1]
    display() 
    
def reverse(event):
    global text_display
    if text_display!="":
        num=float(text_display)
        num=-num
        text_display=str(num)
        display()

def x(event):
    global action
    global text_display
    act=event.widget["text"]
    if act=="%":
        if action!="":
            num=float(text_display)
            num=num/100
            text_display=str(num)
        else:
            text_display=""
        display()
    elif act=="1/x":
        num=float(screen.get())
        if num!=0:
            num=1/num
            text_display=str(num)
            display()
            text_display=""
        else:
            screen.delete(0,"end")
            screen.insert(0,"ERROR")
            text_display=""
    else:
        num=math.sqrt(float(screen.get()))
        text_display=str(num)
        display()
        
dec=False
def decimal(event):
    global dec
    if dec:
        return
    global text_display
    if text_display=="":
        text_display="0"
    text_display+="."
    display()
    
def memory(event):
    global memory
    global text_display
    act=event.widget["text"]
    if act =="MS":
        memory=float(screen.get())
    elif act=="MR":
        text_display=str(memory)
        display()
    elif act=="MR+":
        memory+=float(screen.get())
    elif act=="MR-":
        memory-=float(screen.get())
    else: 
        memory=0
        
        

root=tk.Tk()
root.title("Calculator")
root.geometry("305x390")
root.resizable(0,0)

screen=tk.Entry(root,font=("Helvetica",17),justify="right")
screen.grid(row=0,column=0,columnspan=5,padx=5,pady=5,ipadx=10,ipady=10)
screen.insert(0,"0")

keyboards={"MC":10,"MR":10,"MS":10,"M+":10,"M-":10,"<-":6,"CE":5,"C":4,"+/-":7,"sqrt":9,"7":1,"8":1,"9":1,"/":2,"%":9,"4":1,"5":1,"6":1,"*":2,"1/x":9,
           "1":1,"2":1,"3":1,"-":2,"=":3,"0":1,"":1,".":8,"+":2}

row=1
column=0
keys=list(keyboards.keys())
for i in range(len(keys)):
    if i==26:
        continue
    if i%5==0 and i!=0:
        row+=1
    if i ==24:
        btn=tk.Button(root,text=keys[i],height=5,width=5, font=("Helvetica",10,"bold"))
        btn.grid(row=row,column=i%5,rowspan=2,padx=5,pady=5,ipady=4)   
    elif i==25:
        btn=tk.Button(root,text=keys[i],height=2,width=13, font=("Helvetica",10,"bold"))
        btn.grid(row=row,column=i%5,columnspan=2,padx=5,pady=5)
    else: 
        btn=tk.Button(root,text=keys[i],height=2,width=5, font=("Helvetica",10,"bold"))
        btn.grid(row=row,column=i%5,padx=5,pady=5)
    if keyboards[keys[i]]==1:
        btn.bind("<Button-1>",number_pressed)
    elif keyboards[keys[i]]==2:
        btn.bind("<Button-1>",action_pressed)
    elif keyboards[keys[i]]==3:
        btn.bind("<Button-1>",finish_pressed)
    elif keyboards[keys[i]]==4:
        btn.bind("<Button-1>",clear)
    elif keyboards[keys[i]]==5:
        btn.bind("<Button-1>",clear_e)
    elif keyboards[keys[i]]==6:
        btn.bind("<Button-1>",backspace)
    elif keyboards[keys[i]]==7:
        btn.bind("<Button-1>",reverse)
    elif keyboards[keys[i]]==8:
        btn.bind("<Button-1>",decimal)
    elif keyboards[keys[i]]==9:
        btn.bind("<Button-1>",x)
    else:
        btn.bind("<Button-1>",memory)
       
root.mainloop()
```
